```python
import pandas as pd
import numpy as np
```

```python
from binance.client import Client
import config #config file with api keys
binance = Client(config.binance_api_key, config.binance_api_secret)
```

```python
transactions = pd.read_csv('transactions.csv')
transactions.head()
```

```python
holdings = transactions[['pair','amount']]
holdings = holdings.set_index('pair',drop=True)
holdings = holdings.to_dict()['amount']
holdings
```

```python
prices = binance.get_all_tickers()
prices = { x['symbol']:x['price'] for x in prices }
prices
```

```python
holdings_btc = {}
for sym, amount in holdings.items():
    pair = sym.replace("/","")
    price = float(prices[pair])
    holdings_btc[sym] = amount * price 
holdings_btc
```

```python
holdings_btc['BTC/BTC'] = 0.04/20
NAV = sum(holdings_btc.values())
display(NAV)
```

```python
for k, v in holdings_btc.items():
    delta = v/NAV*100 - 1/20*100
    print(k, '{0:.2f}'.format(delta))
```
